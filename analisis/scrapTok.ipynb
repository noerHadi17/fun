```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import pandas as pd

name = []
seller = []

driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)  # tunggu maksimal 10 detik

for page in range(1, 2):
    url = f"https://www.tokopedia.com/search?navsource=&page={page}&q=rak+susun+serbaguna&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st="
    driver.get(url)

    # Scroll untuk load produk
    for scroll in range(25):
        driver.execute_script("window.scrollBy(0,300)")
        sleep(1)

    # Tunggu elemen seller muncul (menggunakan sebagian class unik)
    try:
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "span[class*='pC8DMVkBZGW7-egObcWMFQ']")))
    except:
        print("Timeout menunggu seller, lanjut...")

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    # Ambil semua produk
    for item in soup.find_all('div', class_="_6+OpBPVGAgqnmycna+bWIw=="):
        # Nama produk
        prod_tag = item.find('span', class_="_0T8-iGxMpV6NEsYEhwkqEg==")
        name.append(prod_tag.get_text() if prod_tag else "No Name Product")

        # Seller
        seller_tag = item.select_one("span[class*='pC8DMVkBZGW7-egObcWMFQ']")
        seller.append(seller_tag.get_text() if seller_tag else "No Name Seller")

driver.quit()

# Buat DataFrame
df = pd.DataFrame({
    'Nama Produk': name,
    'seller': seller
})
df
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
import pandas as pd

# Setup Chrome
options = Options()
options.add_argument('--start-maximized')
options.add_argument('--disable-blink-features=AutomationControlled')  # menghindari deteksi bot
driver = webdriver.Chrome(options=options)

# Buka halaman Tokopedia
url = "https://www.tokopedia.com/search?navsource=&page=1&q=rak+susun+serbaguna"
driver.get(url)
sleep(5)

# Scroll hingga banyak produk termuat
for _ in range(25):
    driver.execute_script("window.scrollBy(0, 500)")
    sleep(1)

# Ambil semua elemen produk
products = driver.find_elements(By.CSS_SELECTOR, "div[data-testid='master-product-card']")

name = []
seller = []

for product in products:
    try:
        # Scroll ke elemen agar seller muncul
        ActionChains(driver).move_to_element(product).perform()
        sleep(0.3)

        # Ambil nama produk
        title = product.find_element(By.CSS_SELECTOR, "div.prd_link-product-name").text
    except:
        title = "No Name Product"

    try:
        # Ambil seller: cari <span> di dalam produk yang mewakili nama toko
        toko = product.find_element(By.CSS_SELECTOR, "span.css-1kr22w3").text
    except:
        toko = "No Name Seller"

    name.append(title)
    seller.append(toko)

driver.quit()

# Buat DataFrame
df = pd.DataFrame({
    'Nama Produk': name,
    'seller': seller
})
df
```
